```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris #load iris
```

```python
#load the dataset
data=load_iris()
```

```python
#identify the target variables(what to classify)
print("Target variables:",data['target_names'])
#unique values
(unique,counts)=np.unique(data['target'],return_counts=True)
print("Unique values are:",unique)
print("Counts are:",counts)
```

```python
sns.barplot(x=data['target_names'],y=counts)
plt.title("Target variable distribution")
plt.show()
```
